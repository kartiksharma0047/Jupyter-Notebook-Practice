```python
# Here We will handle empty or missing data fields
import pandas as pd
csvFile="/WeatherForecastSheet2(Sheet1).csv"
```

```python
def generate_dates(length):
    return [f"{day}/1/2017" for day in range(1, length + 1)]

df = pd.read_csv(csvFile)
df["Day"] = generate_dates(len(df))
df["Day"] = pd.to_datetime(df["Day"], format="%d/%m/%Y")
df
```

```python
# First we need empty value in the csv file so make the day column to index column
df.set_index("Day", inplace=True)
df
```

```python
# (1) fillna()
use_Fill_Na=df.fillna("Nothing Here")
use_Fill_Na
```

```python
# For different null value handling in different columns
use_Fill_Na2=df.fillna({
    "Temperature":0,
    "Windspeed":0,
    "Event":"No Event"
})
use_Fill_Na2
```

```python
# Now if we want to repeat previous day value instead of 0 or any other random value then
use_Fill_Na3=df.fillna(method="ffill")
use_Fill_Na3
```

```python
# If you want backward fill then use bfill method
use_Fill_Na4=df.fillna(method="bfill")
use_Fill_Na4
```

```python
# (2) Interpolate - this will generate a mid value suppose today temp is 32 and day after tommrow temprature is 28 so, Interpolated will give a min value which is 30 on tommorow value
df_Interpolate=df.interpolate()
df_Interpolate
# Will return NaN for string values or values that can't performed mathematical conepts
```

```python
# Now suppose we want a value nearest to the given date then use
df_Interpolate2=df.interpolate(method="time")
df_Interpolate2
```

```python
# (3) dropna() - This will drop all the rows which have no data in it
df_dropna=df.dropna()
df_dropna
```

```python
# If you want to drop a row which is having all na values then do-
df_dropna2=df.dropna(how="all")
df_dropna2
```

```python
# If i have atleast 1 non NaN value then save it -
df_dropna3=df.dropna(thresh=2)
df_dropna3
```
