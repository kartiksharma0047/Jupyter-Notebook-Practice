```python
import pandas as pd

df1=pd.DataFrame({
    "City":["New York","Chicago","Orlando","Sans Fransico"],
    "Temperature":[21,14,35,32]
})
df2=pd.DataFrame({
    "City":["New York","Chicago","Orlando"],
    "Humidity":[68,65,75]
})
```

```python
# I want to join these 2 Data frame to a new data frame which contain both temprature and humidity
df3=pd.merge(df1,df2,on="City")
df3
```

```python
# If we want to include the cities which are not present in df1 and df2 merge then
df3=pd.merge(df1,df2,on="City",how="outer")
df3
```

```python
# We can also use left and right join just like in SQL
df3=pd.merge(df1,df2,on="City",how="left")
df3
```

```python
# Now right join
df3=pd.merge(df1,df2,on="City",how="right")
df3
```

```python
# Now suppose we have 2 data frames having same colums humidity and temperature with different values
new_df1=pd.DataFrame({
    "City":["New York","Chicago","Orlando","Sans Fransico"],
    "Temperature":[21,14,35,32],
    "Humidity":[68,65,75,71]
})
new_df2=pd.DataFrame({
    "City":["New York","Chicago","Orlando"],
    "Temperature":[21,14,35],
    "Humidity":[68,65,75]
})
```

```python
# Now if we join those 2 table
new_df3=pd.merge(new_df1,new_df2,on="City",suffixes=("_left","_right"))
new_df3
```
