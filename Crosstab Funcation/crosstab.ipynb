```python
import pandas as pd

df=pd.read_csv("/Survey_Data(Sheet 1)(Sheet1).csv",sep=";")
df
```

```python
# Lets use crosstab for x axis -> handedness and for Y axis -> Nationality
df2=pd.crosstab(df.Nationality,df.Handedness)
df2
```

```python
# corssTab also returns total count
df3=pd.crosstab(df.Nationality,df.Handedness,margins=True)
df3
```

```python
# Lets try multiple variables
df4=pd.crosstab(df.Nationality,[df.Handedness,df.Sex],margins=True)
df4
```

```python
# To get values in % then
df5=pd.crosstab(df.Nationality,df.Handedness,normalize='index')
df5
```
