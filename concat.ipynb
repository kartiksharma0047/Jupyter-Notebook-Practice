```python
import pandas as pd

india_Weather=pd.DataFrame({
    "City":["Mumbai","Delhi","Bangalore"],
    "Temperature":[32,45,30],
    "Humidity":[80,60,78]
})
us_Weather=pd.DataFrame({
    "City":["New York","Chicago","Orlando"],
    "Temperature":[21,14,35],
    "Humidity":[68,65,75]
})
```

```python
# Concat both weathers
df1=pd.concat([india_Weather,us_Weather],ignore_index=True)
df1
```

```python
# Keys in concat
df2=pd.concat([india_Weather,us_Weather],keys=["India","US"])
df2
```

```python
# To access the subset from the dataframe using keys
df2.loc["US"]
```
