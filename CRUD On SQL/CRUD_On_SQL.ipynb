```python
import pandas as pd
import sqlalchemy as toolkit
```

```python
# Here i am using MySQL so that is why pymysql and sqlalchemy is being used here :
engine=toolkit.create_engine("mysql+pymysql://root:0047@localhost:3306/application")
```

```python
# In my sql schema i have 2 tables - customers and orders
# (1) Customer Table
CustomerTable = pd.read_sql_table("customers",engine)
CustomerTable
```

```python
# (2) Order Table
OrderTable = pd.read_sql_table("orders",engine)
OrderTable
```

```python
# To apply query from pandas to swl table then
query='''
SELECT customers.Name,customers.Phone_Number,orders.Name,orders.Amout
FROM customers INNER JOIN orders
ON customers.ID=orders.Customer_ID
'''
QueryResult=pd.read_sql_query(query,engine)
QueryResult
```

```python
# Now we will append an csv file data to out SQL Table
df=pd.read_csv("Customer_Data(Sheet1)(Sheet1).csv",sep=";")
df
```

```python
# Now lets add data of excel to sql - 
df.to_sql(
    name="customers",
    con=engine,
    index=False,
    if_exists='append'
)
```
