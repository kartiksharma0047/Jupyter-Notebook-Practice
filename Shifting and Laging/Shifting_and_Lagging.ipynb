```python
import pandas as pd
df=pd.read_csv("/Facebook_Stock_Historical_Data.csv",index_col="Date",parse_dates=True)
df.head()
```

```python
# To shift data by 1 row
df.shift(1)
```

```python
# To shift data in reverse
df.shift(-1)
```

```python
# Shift is used to calculate % change of data in n days or calculate the price change in n days example
df["Price Change"]=df["Price"].shift(-1)
df.head()
# So now i get previous days values column
```

```python
df['1 day price change']=df['Price']-df['Price Change']
df.head()
```

```python
# How about calculating 5 days return
df['5 days % return']=(df['Price']-df['Price'].shift(5))*100/df['Price'].shift(5)
df
```
