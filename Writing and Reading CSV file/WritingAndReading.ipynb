```python
# Reading And Writing CSV Files
import pandas as pd

Data='/Compnay_Sheet(Sheet1).csv'
```

```python
# If we want to remove header use (header or skiprows)
pd.read_csv(Data,header=1)
```

```python
# We can also add header by using names or columns
pd.read_csv(Data,names=["hi","hello","bye","lol","No","moveout"])
```

```python
# To select n number of rows
pd.read_csv(Data,nrows=4)
```

```python
# How to make na value to NaN
pd.read_csv(Data,na_values=["n.a.","not available"])
```

```python
pd.read_csv(Data,na_values={
    "Eps":[-1]
})
```

```python
# Writing a File
df=pd.read_csv(Data)
df.to_csv("/new.csv",index=False)
pd.read_csv("/new.csv")
```

```python
# If you want to skip writing headers then
df.to_csv("/new.csv",header=False)
pd.read_csv("/new.csv")
```

```python
pd.read_csv("/new.csv",names=["Tickers","Eps","Revenue","Price","People"])
```

```python
def convert_price_cell(cell):
    if isinstance(cell, str):
        if cell.lower() in ["n.a.", "na", "nan", "none"]:
            return float('nan')  # classic NaN
        elif "$" in cell:
            return cell.strip()
        else:
            return f"${cell.strip()}"
    return cell

newCsv = pd.read_csv(
    "/new.csv",
    names=["Tickers", "Eps", "Revenue", "Price", "People"],
    converters={"Price": convert_price_cell}
)
newCsv
```
