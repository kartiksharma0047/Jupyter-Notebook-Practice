```python
# Working with CSV file with pandas
import pandas as pd
csvFile=pd.read_csv('PracticeForecastFile(Sheet1).csv')
```

```python
# Here the Output is in tabular form
csvFile
```

```python
# Here the output is in text form
print(csvFile)
```

```python
# Some Features of pandas we have 2 fetaure to do the same thing
# (1) Shape :- method(1)
csvFile.shape
```

```python
# (1) Shape :- method(2)
rows, columns=csvFile.shape
print(f"Rows: {rows}")
print(f"Columns: {columns}")
```

```python
# (2) Slicing :- head()
csvFile.head()
```

```python
# (2) Slicing :- tail()
csvFile.tail(2)
```

```python
# (2) Slicing :- indexing
csvFile[2:6]
```

```python
# (3) Printing Column :- example csvFile.event or csvFile['event']
print(csvFile['Day'].head())
```

```python
# To Print Multiple Columns
csvFile[['Day','Temperature']].head(2)
```

```python
# Type of rows and columns
print(type(csvFile[0:1]))
print(type(csvFile['Day']))
```

```python
# (4) Max(), Min(), Mean() and std()
print(f"Maximum Temperature: {csvFile['Temperature'].max()}")
print(f"Minimum Temperature: {csvFile['Temperature'].min()}")
print(f"Mean Temperature: {csvFile['Temperature'].mean()}")
print(f"Maximum Event: {csvFile['Event'].max()}")
print(f"Minimum Event: {csvFile['Event'].min()}")
print(f"Standard Deviation: {csvFile['Temperature'].std()}")
```

```python
# (5) Describe() - It will print all the statastics on the data set
csvFile.describe()
```

```python
# (6) Conditional Statements in DataFrames
# (a) Condition Example
csvFile[csvFile.Temperature>=37]
```

```python
# (b) Print that row where temp is max
csvFile[csvFile.Temperature==csvFile.Temperature.max()]
```

```python
# (c) Print the specific date where temprature is maximum
csvFile['Day'][csvFile.Temperature==csvFile.Temperature.max()]
```
