```python
import pandas as pd
df=pd.read_csv("/WeatherForecast(Sheet3)(Sheet1).csv")
df
```

```python
# Now we use replace method
df1=df.replace(-99999,value="Not Found")
df1
```

```python
# We can also replace multiple values
dd2=df.replace([-99999,"No Event"],value="Not Found")
dd2
```

```python
# Now if we want to replace based on colums then
df3=df.replace({
    "temperature":-99999,
    "windspeed":-99999,
    "event":"No Event"
},value="Not Found"
)
df3
```

```python
# Now we want to map value with specific values
df4=df.replace({
    -99999:"Null Values",
    "No Event":"Not Found"
})
df4
```

```python
# To Chop of specfic values from the data suppose chop 'No' from 'No Event'
df5=df.replace({
    "Event":'No'
    },'',regex=True)
df5
```
