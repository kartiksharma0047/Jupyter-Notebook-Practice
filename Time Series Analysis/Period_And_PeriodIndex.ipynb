```python
# Lets work on data
data={
  "Line Item":["Revenue","Expenses","Profit"],
  "2017Q1":[115904,86544,29360],
  "2017Q2":[120854,89485,31369],
  "2017Q3":[118179,87484,30695],
  "2017Q4":[130936,97743,33193],
  "2018Q1":[117542,87688,29854]
}
df=pd.DataFrame(data)
df
```

```python
# call set_index
df.set_index("Line Item",inplace=True)
df
```

```python
# Now transpose the table
df=df.T
df
```

```python
df.index
# Here we can notice our indexs are of type object and we need to convert into Period Index
```

```python
df.index=pd.PeriodIndex(df.index,freq='Q-JAN')
df.index
```

```python
# Lets create "Start Date" Column
df["Start Date"]=df.index.map(lambda x: x.start_time)
df
```

```python
# Lets create "End Date" Column
df["End Date"] = df.index.map(lambda x: x.end_time.strftime('%Y-%m-%d'))
df
```
