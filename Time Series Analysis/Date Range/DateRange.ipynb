```python
# Date Range in pandas
import pandas as pd

df=pd.read_csv("/Apple_Historical_Data(Sheet1).csv")
df=df[['Close/Last','Volume','Open','High','Low']]
df.head()
```

```python
# Now use pandas Date range function -
range=pd.date_range(start="11/21/2024",end="05/11/2025",freq="B")
range
```

```python
# Now set the date as index into the data frame
df.set_index(range,inplace=True)
df.head()
```

```python
# Lets plot a chart of closinn stock price
%matplotlib inline
df2=df.copy()
df2["Close/Last"]=df2["Close/Last"].str.replace("$","",regex=False).astype('float').astype(int)
df2["Close/Last"].plot()
```

```python
# Generate DF according to the weekends also... Like weekends should have data of last closing day
df3=df2.asfreq('D',method="pad")
df3['Close/Last'].plot()
```
