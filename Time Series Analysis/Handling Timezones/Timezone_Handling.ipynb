```python
# Now here we will handle timezone across different countries
import pandas as pd
df = pd.read_csv("/Google_Stock_Historical_Data.csv", index_col="Date", parse_dates=True)
df['Close/Last'] = df['Close/Last'].str.replace('$', '', regex=False).astype(float).astype(int)
df = df.rename(columns={'Close/Last': 'Price'})
df=df['Price']
df.head()
```

```python
# Lets Examine the index here
df.index
```

```python
# As we can see the date time object is datetime64[ns] means it is not timezone aware object so to convert it
df=df.tz_localize(tz='US/Eastern')
df.index
```

```python
# Lets get berlin timezone
df=df.tz_convert(tz='Europe/Berlin')
df.index
```

```python
# How do we know these string for timezone ex Europe/Berlin or US/Eastern..
# So for that there is a library
from pytz import all_timezones
all_timezones
```

```python
# Now lets covert the df for mumbai time zone which is -
df=df.tz_convert(tz='Asia/Calcutta')
df.index
```

```python
# We can also set Timezone in date_range function ex
rng=pd.date_range(start="1/1/2017",periods=10,freq='h',tz='Asia/Calcutta')
rng
```

```python
# dateutil timezone - it can access the operating system timezone too but pytz have predefined time zone only
rng2=pd.date_range(start="1/1/2017",periods=10,freq='h',tz='dateutil/Asia/Calcutta')
rng2
```
