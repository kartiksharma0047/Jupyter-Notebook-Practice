```python
import pandas as pd

df=pd.read_csv("/Weather_Forecast(Sheet 6)(Sheet1).csv")
df
```

```python
# Lets try to reshape the df using pivot
df1=df.pivot(index="Data",columns="City")
df1
```

```python
# Now if we want specific data to been displayed then use
df2=df.pivot(index="Data",columns="City",values="Humidity")
df2
```

```python
# now we will work with another csv file for pivot table
pivotTable=pd.read_csv("/Weather_Forecast(Sheet 7)(Sheet1).csv")
pivotTable
```

```python
# Lets make it pivot table
pivotTable1=pivotTable.pivot_table(index="Date",columns="City")
pivotTable1
```

```python
# If i want to apply margins then
pivotTable2=pivotTable.pivot_table(index="Date",columns="City",margins=True)
pivotTable2
```

```python
# Now try Grouper in pivot table by new csv file
NewCsv=pd.read_csv("/Weather_Forecast(Sheet 8)(Sheet1).csv")
NewCsv
```

```python
# Lets first convert the string format date into date and time format
NewCsv["Date"]=pd.to_datetime(NewCsv["Date"])
NewCsv
pivotTable3=NewCsv.pivot_table(index=pd.Grouper(freq="ME",key="Date"),columns="City")
pivotTable3
```
